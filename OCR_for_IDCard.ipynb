```python
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt
from PIL import Image

ocr = PaddleOCR(use_angle_cls=True, lang='en')

image_path = '/content/Real Indian Pan Card Images.jpg'
img = Image.open(image_path)
result = ocr.ocr(image_path, cls=True)

detected_texts = [line[1][0] for line in result[0]]

full_text = ' '.join(detected_texts)
print('Detected Text:', full_text)

boxes = [line[0] for line in result[0]]
texts = [line[1][0] for line in result[0]]

img_with_boxes = draw_ocr(img, boxes, texts, font_path='/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf')
plt.figure(figsize=(10, 10))
plt.imshow(img_with_boxes)
plt.axis('off')
plt.show()

```
